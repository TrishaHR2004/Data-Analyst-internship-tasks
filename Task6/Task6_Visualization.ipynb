```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/nkm/Downloads/2015.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Sort top 10 happiest countries
top10 = df.sort_values("Happiness Score", ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top10["Country"], top10["Happiness Score"])
plt.xlabel("Happiness Score")
plt.ylabel("Country")
plt.title("Top 10 Happiest Countries (2015)")
plt.gca().invert_yaxis()
plt.grid(axis='x')

plt.show()

# The top-ranked countries show significantly higher happiness scores,
#indicating a strong relationship between socio-economic stability and overall well-being. 
#The score gap between the 1st and 10th ranked country highlights how small changes in GDP, health, and freedom can influence national happiness rankings.
```

```python
# Sort by GDP for smooth trend line
gdp_trend = df.sort_values("Economy (GDP per Capita)")

plt.figure(figsize=(10,6))
plt.plot(
    gdp_trend["Economy (GDP per Capita)"],
    gdp_trend["Happiness Score"]
)

plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("Trend: GDP per Capita vs Happiness Score (2015)")
plt.grid(True)

plt.show()

#The upward trend indicates a positive correlation between GDP per capita and happiness score. While economic growth improves happiness, 
#the curve flattens at higher GDP levels, suggesting diminishing returns beyond a certain income threshold.
```

```python
plt.figure(figsize=(10,6))

plt.hist(df["Happiness Score"], bins=15)
plt.xlabel("Happiness Score")
plt.ylabel("Number of Countries")
plt.title("Distribution of Happiness Scores (2015)")
plt.grid(True)

plt.show()

#The histogram shows that most countries fall within a mid-range happiness score, 
#while only a few countries appear as outliers at the high and low ends. 
#This indicates global happiness is generally concentrated around an average level, with limited extreme cases.
```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Economy (GDP per Capita)"],
    df["Happiness Score"]
)

plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.title("Correlation: GDP vs Happiness Score (2015)")
plt.grid(True)

plt.show()


#"This scatter plot shows a positive correlation between GDP per capita and happiness score.
#As GDP increases, happiness generally rises, though some countries deviate, 
#indicating non-economic factors like freedom and trust also influence happiness."
```
